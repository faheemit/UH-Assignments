```python
# Define the points
x1, y1 = 3, 5
x2, y2 = 7, 2

plt.figure(figsize=(8, 6))

# Plot the points
plt.scatter([x1, x2], [y1, y2], color='black', s=100, label='Points')

# Plot the line passing through the points and extending beyond
plt.plot([x1, x2], [y1, y2], color='black', linewidth=1, linestyle='-', label='Line')

# Extend the line beyond the points
plt.plot([0, 10], [5 + (0 - x1) * (y2 - y1) / (x2 - x1), 5 + (10 - x1) * (y2 - y1) / (x2 - x1)], color='black', linewidth=1, linestyle='-')

# Set appropriate ranges for both the x and y axes
plt.xlim(0, 10)
plt.ylim(0, 8)

# Set labeled and legible axes
plt.xlabel('X', fontsize=14)
plt.ylabel('Y', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add a coordinate grid
plt.grid(True)

# Add legend
plt.legend(fontsize=14)

# Show the plot
plt.show()
```

```python
x = df['var1']
y = df['var2']

# Create scatter plot
plt.scatter(x, y, c='red', alpha=0.5)  # 'c' specifies the color, 'alpha' controls transparency
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)  # Add gridlines
plt.show()
```

```python
# Choose multiple columns for scatter plot
x_columns = ['var1', 'var2', 'var3', 'class']  # Choose the columns for the X-axis
y_columns = ['var1', 'var2', 'var3','class']  # Choose the columns for the Y-axis

# Create scatter plot
fig, axes = plt.subplots(nrows=len(y_columns), ncols=len(x_columns), figsize=(20, 20))

for i, y_col in enumerate(y_columns):
    for j, x_col in enumerate(x_columns):
        ax = axes[i, j]
        ax.scatter(df[x_col], df[y_col], color = 'red')
        ax.set_title('{} vs {}'.format(y_col, x_col))
        ax.set_xlabel(x_col)
        ax.set_ylabel(y_col)
        ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
columns = df[1:]
# Create histograms for each column combine
plt.figure(figsize=(12, 8))
for column in columns.columns:
    plt.hist(df[column], bins=10, alpha=0.5, label=column)

plt.title('Histogram of All Columns')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#violinplot of all columns

fig, axes = plt.subplots(nrows=1, ncols=len(df.columns), figsize=(12, 6))

for i, column in enumerate(df.columns):
    axes[i].violinplot(df[column])
    axes[i].set_title('Violin plot of {}'.format(column))
    axes[i].set_ylabel(column)
    for pc in axes[i].collections:
        pc.set_facecolor('green')
        
plt.tight_layout()
plt.show()
```
